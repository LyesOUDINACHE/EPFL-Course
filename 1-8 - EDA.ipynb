```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
X = np.linspace(-16, 16)
Y1 = np.sin(X) / X
Y2 = 2 * np.sin(X) / X
Y3 = 3 * np.sin(X) / X
plt.plot(X,Y1, color = 'blue', linestyle = '--', linewidth = 2, label = '$\sin(x)/x$');
plt.plot(X,Y2, color = 'green', linestyle = '-.',linewidth = 2, label = '$2*\sin(x)/x$');
plt.plot(X,Y3, color = 'red', marker = 'o',linewidth = 2, label = '$3*\sin(x)/x$');
plt.ylim(-1,3.5);
plt.yticks([-1,0,1,2,3]);
plt.legend()
plt.show()
```

If matplotlib tries to make easy things easy and hard things possible, Seaborn tries to make a well-defined set of hard things easy too.

Seaborn is not meant as a replacement to matplotlib, but more as a complimentary packaged. Some of its main benefits are

default aesthetics are much more appealing
easier to customize the plots

```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.distplot(iris['sepal_length']);
```

```python
sns.regplot(iris.sepal_length, iris.petal_length)
plt.title("A Title") #Add plot title
plt.ylabel("y-label") #Adjust the label of the y-axis
plt.xlabel("x-label") #Adjust the label of the x-axis
plt.ylim(0,10) #Adjust the limits of the y-axis
plt.xlim(0,10) #Adjust the limits of the x-axis
sns.set_style('whitegrid')
plt.show()
```


#### Pairplot
Seaborn has a very useful method called pairplot(). What this function does is that it takes each numerical variable of our DataFrame, and then creates a grid of scatter plots for each pairwise relationship

```python
sns.pairplot(iris);
```

```python
sns.pairplot(iris, kind='reg');
```

This is extremely useful for exploring correlations between multi-dimensional data, such as in our example here. For example, we can right away notice an almost linear relationship between the petal length and petal width. We will look more at correlation and its meaning in a future subject.

If we then want to investigate further just one of the scatter plots, we can specify this in the arguments to the pairplot() method as follows:

```python
sns.pairplot(iris, x_vars = ['sepal_length'], y_vars=['petal_length'], height = 10);
```

```python
sns.pairplot(iris, x_vars=['sepal_length'], y_vars = ['petal_length'], height = 10, hue='species', palette = 'pastel')
plt.title('The Iris data set: petal length vs sepal length');
## beaucoup plus simple que matplotlib et palette de couleurs
```

```python
my_palette=["#95a5a6", "#e74c3c", "#34495e"]
sns.pairplot(iris, x_vars=['sepal_length'], y_vars = ['petal_length'], height=10, hue='species', palette=my_palette);
```

#### FacetGrid
In addition to pairplot(), perhaps the second most useful feature of Seaborn is the ability to build conditional plots. What exactly is a conditional plot? It's a plot where the data is divided by one or more variables. It is usually used to obtain multiple plots side by side where each plot corresponds to a certain subset of the data.

```python
grid = sns.FacetGrid(iris, col='species')
grid = grid.map(plt.scatter, "sepal_length", "petal_length");
```

#### Violon Plots and Swarm Plots

```python
my_palette = ["pink","lightgreen","lightblue"]
sns.violinplot(x=iris.species, y=iris.petal_length, palette=my_palette);
plt.title("A violin plot of petal length");
```

```python
my_palette = ["pink","lightgreen","lightblue"]
sns.swarmplot('species', 'sepal_length', data = iris, palette = my_palette);
plt.title("A swarm plot of sepal length");
```
