Exercise - Load and plot data
---

> **Exercise**: Load the data and plot the `x1` and `x2` variables. Use one color for each class.

```python
import pandas as pd
import numpy as np

# Load data
data_df = pd.read_csv('spirals.csv')

# First five rows
data_df.head()
```

```python
# Create X/y arrays
X = data_df.drop('y', axis=1)
y = data_df['y']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Same scale for x- and y-axis
fig, ax = plt.subplots()
ax.set_aspect('equal', adjustable='box')

# Plot data
class1_idx = (y == 1)
ax.scatter(X['x1'][class1_idx], X['x2'][class1_idx], c='C3', label='class 1', s=4)
ax.scatter(X['x1'][~class1_idx], X['x2'][~class1_idx], c='C0', label='class 0', s = 4)
ax.legend()
plt.show()
```

```python


# Helper function
def decision_surface(ax, x1, x2, y, estimator):
    # Same scale for x- and y-axis
    ax.set_aspect('equal', adjustable='box')

    # Plot data points
    class1_idx = (y == 1)
    plt.scatter(x1[class1_idx], x2[class1_idx], color='C3', label='class 1', s = 4)
    plt.scatter(x1[~class1_idx], x2[~class1_idx], color='C0', label='class 0', s = 4)

    # Create a grid of values
    xlim, ylim = ax.get_xlim(), ax.get_ylim()
    x_values = np.linspace(*xlim, num=500)
    y_values = np.linspace(*ylim, num=500)
    xx, yy = np.meshgrid(x_values, y_values)
    grid_points = np.c_[xx.flatten(), yy.flatten()]

    # Compute predictions
    preds = estimator.predict(grid_points)
    zz = preds.reshape(xx.shape)

    # Draw decision boundary
    plt.contour(xx, yy, zz, levels=[0.5], colors='gray')

    # Plot decision surface with level curves
    plt.contourf(xx, yy, zz, alpha=0.1, cmap=plt.cm.coolwarm)
    plt.legend()
    plt.show()
```

### Logistic Regression 



```python
from sklearn.model_selection import train_test_split

# Split into train/test sets
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.3, random_state=0)

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Create the estimator
logreg = make_pipeline(StandardScaler(), LogisticRegression())

# Fit it to train data
logreg.fit(X_tr, y_tr);
```

```python
# Plot decision surface
fig, ax = plt.subplots()
decision_surface(ax, X['x1'], X['x2'], y, logreg)
```

### SVM with linear Kernel


```python
from sklearn.svm import LinearSVC

# Create SVM with linear kernel
linear_svc = LinearSVC()

# Fit estimator
linear_svc.fit(X, y)
```

```python
# Plot decision surface
fig, ax = plt.subplots()
decision_surface(ax, X['x1'], X['x2'], y, linear_svc)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Create a k-NN pipeline
knn_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=4))
])

# Fit estimator
knn_pipe.fit(X_tr, y_tr)
```

```python
# Plot decision surface
fig, ax = plt.subplots()
decision_surface(ax, X['x1'], X['x2'], y, knn_pipe)
```

### Decision Tree

```python
from sklearn.ensemble import RandomForestClassifier

# Create a decision tree
dt = RandomForestClassifier(
    n_estimators=1, max_depth=10, random_state=0)

# Fit estimator
dt.fit(X, y)
```

```python
# Plot decision surface
fig, ax = plt.subplots()
decision_surface(ax, X['x1'], X['x2'], y, dt)
```

### Random Forest

```python
# Create a (large) random forest estimator
rf_200 = RandomForestClassifier(
    n_estimators=200, max_depth= None, random_state=0)

# Fit estimator
rf_200.fit(X, y)

# Plot decision surface
fig, ax = plt.subplots()
decision_surface(ax, X['x1'], X['x2'], y, rf_200)
```

### SVM with RBF kernel

```python
from sklearn.svm import SVC

# Create SVM with RBF kernel
rbf_svc = SVC(kernel='rbf', C=1, gamma=1)
# Fit estimator
rbf_svc.fit(X, y)

# Plot decision surface
fig, ax = plt.subplots()
decision_surface(ax, X['x1'], X['x2'], y, rbf_svc)
```
